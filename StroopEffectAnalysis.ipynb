# Stroop Effect Analysis

## 1. Background Information

In a Stroop task, participants are presented with a list of words, with each word displayed in a color of ink. The participantâ€™s task is to say out loud the color of the ink in which the word is printed. The task has two conditions: a congruent words condition, and an incongruent words condition. In the congruent words condition, the words being displayed are color words whose names match the colors in which they are printed: for example RED, BLUE. In the incongruent words condition, the words displayed are color words whose names do not match the colors in which they are printed: for example PURPLE, ORANGE. In each case, we measure the time it takes to name the ink colors in equally-sized lists. Each participant will go through and record a time from each condition.

## 2. Objective

To determine if the time that the participant takes to read or the reaction time (RT) the color list on the a incongruent words condition (IC) is higher on average than in the congruent condition (CC).

## 3. Method

To verify if RT is significant longer on average, I've performed an upper-tailed t-test at a Alpha level $\alpha$ of .01 for our two paired samples.

The independent variable on this experiment is the condition beeing congruent (CC) or incongruent (IC). The dependent variable is the time someone takes to read the colors on each condition. Our null hyphoteses is that RT on IC is not higher on average and our alternative hyphotesis is that RT is higher on IC on average. i.e.

\begin{align}
H_{0}: \mu_{A} - \mu_{B} = 0
\end{align}
\begin{align}
H_{a}: \mu_{A} - \mu_{B} \neq 0
\end{align}

Where $\mu_{A}$ and $\mu_{B}$ are the average RT on each condition (CC and IC).

If our t-statistic is outside the quantiles of the t-distribution corresponding to our confidence level and degrees of freedom (df), we will <u>reject</u> the null hypothesis. 


## 4. Computation

### 4.1. Setup

```python
import numpy as np
import pandas as pd
import scipy
import scipy.stats as stats
from IPython.display import display
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')

data = pd.read_csv('stroopdata.csv', sep=',')
```

### 4.2. Data Visualization

```python
data.boxplot(return_type='axes')
```

Box plot shows significant diferences bettwen RT sample distributions, I would expect it to be significantly higher on average since the lowest observation at the IC distribution is above the median of the CC distribution. $^{[1]}$

```python
data.plot(kind='density', subplots=False, sharex=False)
```

### 4.3. Statistics Sumary:

```python
ttr_diff = pd.DataFrame({"ttr_before":data.Congruent,
                          "ttr_after":data.Incongruent,
                          "ttr_diff":data.Incongruent-data.Congruent})
ttr_diff.describe()
```

Above summary shows that RT was 7.96 seconds higher on average on the incongruent condition. Standard deviation for differences is 4.86.

Let's conduct a paired t-test to see whether this difference is significant at a 99% confidence level$^{[2]}$:

### 4.4. Finding t-statistic

```python
scipy.stats.ttest_rel(a = data.Incongruent, b = data.Congruent)
```

Our t-statistic is equal to 8.02 $^{[3]}$


### 4.5. Finding t-critical

Since we have 24 observations our df is 23 (n-1).

Checking our upper-tailed t-critical at $\alpha$=.01 with stats.t.ppf():

```python
scipy.stats.t.ppf(q=0.99, df=23)
```

### 4.6. Statistic Decision

Since our t-statistic lies way above our t-critical at $\alpha$=.01 we <u>reject</u> the null hyphoteses.


## 5. Conclusion

Our test shows that participants on the incongruent condition took statistically significant higher times on average. We can conclude at a 99% confidence interval that this results are not due to chance.

Those results are expected since reading is a automated process, where the mind automatically determines the semantic meaning of the word (it reads the word "red" and thinks of the color "red"), and then must intentionally check itself and identify instead the color of the word (the ink is a color other than red), a process that is not automated. $^{[4]}$

Similar effect would be found if participants where requested to name objects or animals by observing their pictures with names writen bellow then in congruent and incongruent conditions. In another variation the participant must name the color of the word instead of the emotional word. $^{[5]}$
